```python
class Atm:
    def __init__(self):
        self.balance = 10000
        self.pincode = '0000'
        self.menu()

    def menu(self):
        print(
            ''' 
           Hi, How would you like to proceed?
           1. Enter 1 to create pin
           2. Enter 2 to deposite
           3. Enter 3 to widraw
           4. Enter 4 to check balance
           5. Enter 5 to exit
'''
        )
        
        user_input= int(input('Select a number!'))
    
        if(user_input==1):
            self.create_pin()
        elif(user_input==2):
            self.deposit_money()
        elif(user_input==3):
            self.widraw_money()
        elif(user_input==4):
            self.check_balance()
        else:
            print('exit')
    def create_pin(self):
         self.pincode = input('Enter your pin')
         print('Pin is set successfully!')
    def deposit_money(self):
        temp = input('Enter your pin')
        if(temp==self.pincode):
            depositeMoney = int(input('Enter your amount'))
            self.balance = self.balance + depositeMoney
            print('Money deposite Successfully')
        else:
            print("invalid pin")
    def widraw_money(self):
        temp = input('Enter your pin')
        if(temp==self.pincode):
            widrawMoney = int(input('Enter your amount'))
            if(widrawMoney<self.balance):
                self.balance = self.balance - widrawMoney
                print('Please take your amount!')
            else:
                print('insufficient funds')
        else:
            print("invalid pin")
    def check_balance(self):
        temp = input('Enter your pin')
        if(temp==self.pincode):
            print(self.balance)
        else:
            print("invalid pin")





```

```python
class Customer():
    __name =''
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address
    def __str__(self):
        return f'My name is in {self.name}. I live {self.address.city} city, pincode is {self.address.pin}, state is {self.address.state}'
        
    def edit_profile(self,new_name,new_state,new_pin,new_city):
        self.name = new_name
        self.address.change_address(new_city,new_pin,new_state)
        

class Adress():
    def __init__(self,city,pin,state):
        self.city = city
        self.pin = pin
        self.state = state
    def __str__(self):
        return f'I live in {self.city} city, pincode is {self.pin}, state is {self.state}'
    
    def change_address(self,new_city,new_pin,new_state):
        self.city = new_city
        self.pin = new_pin
        self.state = new_state

add = Adress('Faisalabad',38000,'Punjab')
print(add)
cust = Customer('Aaqib','Male',add)
print(cust)

cust.edit_profile('Zohaib','Punjab','38000','Faisalabad')
print(cust)


        
        
```
