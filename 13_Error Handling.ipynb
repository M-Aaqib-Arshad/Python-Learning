### Assert 
assert in Python is used to check that a condition is true during program execution.

If the condition is True, the program continues.

If the condition is False, Python raises an AssertionError.

```python
x = 10
assert x > 0, 'Condition unsatisfied!'
print('Condition is true continues.')
```

```python
x = 10
assert x < 0, 'Condition unsatisfied!'
print('Condition is true continues.')
```
