## Sets
A set in Python is an unordered collection of unique elements.

**Why is the order different?**

Because sets do not preserve insertion order for iteration in a guaranteed way. Python stores set elements using a hash table, so elements are arranged based on their hash values rather than the order you added them.

General Rule
List → ordered collection
Tuple → ordered collection
Set → unordered collection of unique items
Dictionary → key-value mapping (preserves insertion order in modern Python)
